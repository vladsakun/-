**–°–∞–∫—É–Ω–∞ –í–ª–∞–¥–∏—Å–ª–∞–≤–∞ –û–ª–µ–∫—Å–∞–Ω–¥—Ä–æ–≤–∏—á–∞**

–¢–µ–º–∞: –ø–æ–±—É–¥–æ–≤–∞ –¥–≤–æ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–æ—ó RNN –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó —ñ–∑ —à—É–º–æ–º

**–ó–∞–≤–¥–∞–Ω–Ω—è**

–ü–æ–±—É–¥—É–≤–∞—Ç–∏ –¥–≤–æ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω—É RNN –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó —ñ–∑ —à—É–º–æ–º. –ü—ñ–¥—ñ–±—Ä–∞—Ç–∏ –≤–ª–∞—Å–Ω—É –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É, –∫—ñ–ª—å–∫—ñ—Å—Ç—å —à–∞—Ä—ñ–≤, —Ñ—É–Ω–∫—Ü—ñ—ó –∞–∫—Ç–∏–≤–∞—Ü—ñ—ó. –ü–æ–±—É–¥—É–≤–∞—Ç–∏ –≥—Ä–∞—Ñ—ñ–∫. N = 10000 data = np.array([np.—Ñ—É–Ω–∫—Ü—ñ—è for x in range(N)]) + 0.1*np.random.randn(N) –í —è–∫–æ—Å—Ç—ñ —Ñ—É–Ω–∫—Ü—ñ—ó –≤–∑—è—Ç–∏ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥ ùëêùëúùë†(3ùë•) , ùë†ùëñùëõ(ùë• ‚àï 5) , ùë†ùëñùëõ(ùë• + 3) ‚àï 2 —ñ —Ç–¥.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

N = 10000
X = np.random.rand(N) * 10
Y = np.cos(3*X) + 0.1 * np.random.randn(N)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, input_shape=(1,), activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse')
history = model.fit(X, Y, epochs=10, batch_size=32, verbose=1)
predictions = model.predict(X)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X, Y, label='–î–∞–Ω—ñ')
plt.scatter(X, predictions, color='r', label='–ü—Ä–æ–≥–Ω–æ–∑–∏')
plt.title('–°–ø—ñ–≤—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –î–∞–Ω–∏—Ö —Ç–∞ –ü—Ä–æ–≥–Ω–æ–∑—ñ–≤')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title('–§—É–Ω–∫—Ü—ñ—è –í—Ç—Ä–∞—Ç')
plt.xlabel('–ï–ø–æ—Ö–∏')
plt.ylabel('–í—Ç—Ä–∞—Ç–∏')

plt.show()
```
